##**Data analysis with groupby and Visualization**

```python
yearsuicide = su.groupby('year')['suicide_rate'].sum()
yearsuicide
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(y = yearsuicide.values, x = yearsuicide.index, color = 'red')
plt.title ("Trends in Suicide Rates (1985-2016)", fontsize = 14, fontweight ='bold', color = 'black')
plt.xlabel("Year", fontsize =11, color = 'black', fontweight='bold')
plt.ylabel("Suicide rate per 100k population", fontsize =11, color = 'black', fontweight='bold')

plt.show()
```

```python
countrysuicide = su.groupby('country')['suicide_rate'].sum().sort_values(ascending = False).head(5)
countrysuicide
```

```python
plt.figure(figsize=(10,6))
sns.barplot(y = countrysuicide.values, x = countrysuicide.index, color = 'blue')
plt.title ("Top 5 Countries with the Highest Rates of Suicide", fontsize = 14, fontweight ='bold')
plt.xlabel("country", fontsize =11, color = 'black', fontweight='bold')
plt.ylabel("Suicide rate per 100k population", fontsize =11, color = 'black', fontweight='bold')


plt.show()
```

```python
country_suicide = su.groupby('country')['suicide_rate'].sum().sort_values(ascending = False).tail(5)
country_suicide
```

```python
plt.figure(figsize=(10,6))
sns.barplot(y = country_suicide.values, x = country_suicide.index, color = 'green')
plt.title ("Top 5 Countries with the Least Rates of Suicide", fontsize = 14, fontweight ='bold')
plt.xlabel("country", fontsize =11, color = 'black', fontweight='bold')
plt.ylabel("Suicide rate per 100k population", fontsize =11, color = 'black', fontweight='bold')

plt.show()
```
